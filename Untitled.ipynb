```python
#Cleaning up the data
#check for duplicated rows
dupe=df.duplicated()
df.loc[dupe]
```

```python
#PRICE has 98216 non-NA values. Since there are 158957 rows, there are many rows without a price. 
#Create a new dataframe eliminating those null rows
df = df.dropna(axis=0, subset=['PRICE'])
```

```python
len (df)
```

```python
#OMG - it worked!!!!
df['PRICE'].describe()
```

```python
#further reduce table to rows where price is between 25% and 75% percentil of values -- ie,, the middle 50% of prices
midprices = df.loc[(df['PRICE'] > 240000) & (df['PRICE'] < 652000)]
midprices
```

```python
#Check to see the difference in the maximum number of bathrooms for the whole data set vs. for the dataset containing the middle
# 50% of prices
midprices['BATHRM'].max()
```
